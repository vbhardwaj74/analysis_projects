```python
#callbacks.ipynb
@app.callback(
    Output('our_graph1','figure'),
    [Input('rating','value'),
    Input('is_satisfied','value'),
    Input('months', 'value')], prevent_initial_call=True
)

def build_graph(rating_val, sat_val, month_val):
    dff=df1[(df1['rating']==rating_val)|
           (df1['is_satisfied']==sat_val)|
           (df1['months']==month_val)]
    # print(dff[:5])

    fig = px.pie(dff, values='visit_id', names='source', height=600)
    fig.update_layout(yaxis={'title':'Visits'},
                      title={'text':'MoM Visits by Source',
                      'font':{'size':28},'x':0.5,'xanchor':'center'})
    return fig

@app.callback(
    Output('our_graph2','figure'),
    [Input('mobile_devices','value'),
     Input('is_satisfied','value'),
     Input('months', 'value')], prevent_initial_call=True
)

def build_graph(mobile_devices_val, sat_val, month_val):
    dff=df2[(df2['mobile_devices']==mobile_devices_val)|
           (df2['is_satisfied']==sat_val)|
           (df2['months']==month_val)]
    # print(dff[:5])

    fig = px.pie(dff, values='visit_id', names='source', height=600)
    fig.update_layout(yaxis={'title':'Visits'},
                      title={'text':'MoM Visits by Mobile Platform',
                      'font':{'size':28},'x':0.5,'xanchor':'center'})
    return fig

@app.callback(
    Output('our_graph3','figure'),
    [Input('parent_desktop','value'),
     Input('is_satisfied','value'),
     Input('months', 'value')], prevent_initial_call=True
)

def build_graph(desktop_val, sat_val, month_val):
    dff=df3[(df3['parent_desktop']==desktop_val)|
           (df3['is_satisfied']==sat_val)|
           (df3['months']==month_val)]
    # print(dff[:5])

    fig = px.pie(dff, values='visit_id', names='source', height=600)
    fig.update_layout(yaxis={'title':'Visits'},
                      title={'text':'MoM Visits by Desktop Platform',
                      'font':{'size':28},'x':0.5,'xanchor':'center'})
    return fig
```

```python
#index.ipynb
app.layout = html.Div([
    html.Div([
        dcc.Link('Source|', href='/apps/source'),
        dcc.Link('Mobile Platforms|', href='/apps/mobile_platforms'),
        dcc.Link('Desktop Platforms', href='/apps/desktop_platforms')
    ], className="row"),
    dcc.Location(id='url', refresh=False),
    html.Div(id='page-content', children=[])
])

@app.callback(Output('page-content', 'children'),
              [Input('url', 'pathname')])
def display_page(pathname):
    if pathname == '/apps/source':
        return layout1
    if pathname == '/apps/mobile_platforms':
        return layout2
    if pathname == '/apps/desktop_platforms':
        return layout3
    else:
        return "Please choose a link"


if __name__ == '__main__':
    app.run_server(debug=False)
```
